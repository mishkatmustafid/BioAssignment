# 2. Find most frequent pattern of a particular length from text

```python
str1 = input()
k = int(input())
a = len(str1) - k +1

ans = []
var = []
var1 = []

for i in range(0,a):
    str2 = ""
    str3 = ""
    for j in range(i,i+k):
        
        str2 = str1[j]
        str3 = str3+str2
    ans.append(str3)

for i in range(0,len(ans)):
    count = 0
    for j in range(i,len(ans)):
        
        if(ans[i] == ans[j]):
            count+=1;
        else:
            count = count
    var.append(count)
t=-1
for i in range(0,len(var)):
    t = max(t,var[i])
    
for i in range(0,len(var)):
    if(var[i] == t):
        var1.append(i)
        
for i in range(0,len(var1)):
    print(ans[var1[i]])
```
